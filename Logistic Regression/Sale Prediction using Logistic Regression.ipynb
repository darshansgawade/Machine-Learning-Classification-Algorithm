### Creating Subsets such as X, Y

```python
X = data.iloc[:,:-1]
X
```

```python
Y = data.iloc[:,-1]
Y
```

### Splitting the X and Y using train_test_split function

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
x_train
```

```python
y_train
pd.DataFrame(y_train)
```

```python
x_test
```

```python
y_test
pd.DataFrame(y_test)
```

### Feature Scaling using Standard Scalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
#fit_transfrom means it is method calculating mean and variance of each of the feature present in the dataset
x_train = sc.fit_transform(x_train)
x_train
```

```python
#transform means transformnig all the features with respective mean and variance
# This is done to make sure that our data should be completely new for the model
x_test=sc.transform(x_test)
x_test
```

### Training The Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

### Prediction 

```python
age = int(input("Enter the Age : "))
salary = int(input("Enter the Salary : "))
newcust = [[age,salary]]
result = model.predict(sc.transform(newcust))
print(result)
if(result==1):
    print("Customer will buy")
else:
    print("Customer will not buy")
```

### Predicting for all values


```python
y_pred = model.predict(x_test)
y_pred
pd.DataFrame(y_pred)
```

### Validation

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
print("Confusion Matrix is : ")
print(cm)
```

```python
precision = precision_score(y_test,y_pred)
print("Precision Score is :")
print(precision)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy Score is :")
print(accuracy)
```

```python
recall = recall_score(y_test,y_pred)
print("Recall Score is :")
print(recall)
```
